```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.linear_model import LinearRegression, Lasso, Ridge
from sklearn.svm import SVR
from xgboost import XGBRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
def evaluate_models(y,y_pred):
    mae = mean_absolute_error(y, y_pred)
    mse = mean_squared_error(y, y_pred)
    rmse = np.sqrt(mse)
    r2 = r2_score(y, y_pred)
    return {'MAE': mae, 'MSE': mse, 'RMSE': rmse, 'R2': r2}
```

```python
def model(name,model):
    model.fit(X_train.drop(columns=['Price_Per_Year_Old']), y_train)
    y_pred_test = model.predict(X_test.drop(columns=['Price_Per_Year_Old']))
    y_pred_train = model.predict(X_train.drop(columns=['Price_Per_Year_Old']))
    results_train[name] = evaluate_models(y_train,y_pred_train)
    results_test[name] = evaluate_models(y_test,y_pred_test)
```

#SHAP for Global Explaination (explains how each feature in dataset affects  model’s predictions)

```python
import shap
import joblib

# Save the XGBoost model
best_model = XGBRegressor()
best_model.fit(X_train, y_train)
joblib.dump(best_model, "best_model.pkl")

# Save the processed data
X.to_csv("processed_data.csv", index=False)

# Load the saved model and data
model = joblib.load("best_model.pkl")
X_loaded = pd.read_csv("processed_data.csv")

# SHAP explainer (works well with tree models)
explainer = shap.Explainer(model, X_loaded)
shap_values = explainer(X_loaded)

# Global feature importance
shap.summary_plot(shap_values, X_loaded)

# Local explanation for one prediction
shap.force_plot(explainer.expected_value, shap_values[0].values, X_loaded.iloc[0,:])
```

If the dot is to the left (< 0) → that feature reduces the predicted price.

If the dot is to the right (> 0) → that feature increases the predicted price.

```python
#Bar chart to show feature importance


# Step 1: compute mean(|SHAP|) for each encoded  column
importance_df = pd.DataFrame({
    "feature": X.columns,
    "importance": np.abs(shap_values).mean(axis=0)
})

# Step 2: group back by original feature name (split by "_" if one-hot encoded)
importance_df["grouped_feature"] = importance_df["feature"].str.split("_").str[0]

# Step 3: aggregate (sum or mean)
grouped_importance = importance_df.groupby("grouped_feature")["importance"].sum().reset_index()

# Step 4: sort values
grouped_importance = grouped_importance.sort_values("importance", ascending=False)

# Step 5: plot
plt.figure(figsize=(10,6))
plt.barh(grouped_importance["grouped_feature"], grouped_importance["importance"], color="skyblue")
plt.xlabel("Mean |SHAP value| (feature importance)")
plt.title("Overall Feature Importance (Grouped by Original Feature)")
plt.gca().invert_yaxis()
plt.show()
```

**Because the car is relatively new (Car_Age effect), its price is high, but based on the  metric like Mileage, I also think it should be lower. After adding all contributions, I arrive at the final predicted price**
